### Categorical Columns Analysis

```python
# Examining the distribution of unique categories in categorical columns
categorical_cols = df[['iaid', 'brid','genderid', 'amount_of_creditcards', 'nationality', 'occupation', 'fee_structure', 'loyalty_classification', 'properties_owned', 'risk_weighting', 'Income Band']].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

### Numerical Columns Analysis

```python
numerical_cols = ['estimated_income', 'superannuation_savings', 'creditcard_balance', 'bank_loans', 'bank_deposits', 'checking_accounts', 'saving_accounts','foreigncurrency_account','business_lending']

plt.figure(figsize = (20,15))
for i,col in enumerate(numerical_cols):
    plt.subplot(3,3,i+1)
    sns.histplot(df[col],kde= True)
    plt.show
```

```python
# HEAT MAP

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(15,15))
sns.heatmap(correlation_matrix, annot = True, linewidth=.5 ,fmt= '.2f' )
```
